```python
#Written by MrZiemni0k // PART 1
#My 1st WebScraping Project. Still needs some improvement but at least it works.
#Should add re.compile as argument for func geturlImgPage - maybe next time :)

import requests
import bs4
import re
from urllib.error import HTTPError
from urllib.error import URLError
import time


#The Archive Page is 3-page deep(Main page, Sub page(Category) and threads with content. Typical forum.
#The main goal for the 1st part of the script is to store all img urls hidden within bbcode(plain text).
#Subpage urls stored in mainlist, threads in sublist and img urls in imglist.

# Function to store thread urls in sublist.
# Passed argument will be url for subpage stored in mainlist.
def geturlThreadPage(pageUrl):
    
    n = 1                   #Page number to jump futher into subpage.
    while True:             #Loop for increasing page number within subpage url with the same pattern .../{mainlist}-page-{number}.html
        try:                #Trying to avoid error in the middle of script if there is any problem with connection.
            
            website = 'http://www.maya-club.com/{}-page-{}.html'.format(pageUrl[:-5],n)  #Slicing {mainlist} to put "-page-{number}"
            newweb = requests.get(website)      #Getting info about the website. 
            print(website)                      #Keeping track of the website url/num.
            time.sleep(2.5)                     #Giving a rest for a server - Good, gut bot :3
        except HTTPError: 
            return("HTTPError")
        except URLError:
            return('The server could not be found!')
             
        bs = bs4.BeautifulSoup(newweb.text,"lxml")        #Converting {newweb} into readable form of html.
        looplist = []                           #Empty list for breaking the loop if url is not found on empty page. 
                                                #Let's say subpage count is 50, so on 51 I want the loop to stop.
        
        for url in bs.find_all("li"):                     #Urls are stored in tables - looking there.
            if url.a != None:                             #If urls stored at "a href=" withing tables:
                sublist.append(url.a["href"])             #Add urls to sublist for the next func to work with later.
                looplist.append(url.a["href"])            #Add urls to looplist to chck later if any urls found within bs.find_all.
        if len(looplist) == 0:                            #If looplist is empty/no stored urls then most probably we chcecked all pgs within subpage.
            print("Breaking loop at{}".format(website))   #Giving info to check progress.
            time.sleep(2)                                 #Being gud, gut bot :3
            break
        else:
            n += 1                                        #Urls found make n(page)+1

# Function to store image_urls in imglist.
# Passed arguments are:
#                     url for thread stored in sublist
#                     a list to store our imgurls. 
def geturlImgPage(urlsub,imglist):
    
    n = 1            #Page number to jump futher into subpage.
    while True:      #Loop for increasing page number within thread url with the same pattern .../{sublist}-page-{number}.html
        try:         #Trying to avoid error in the middle of script if there is any problem with connection.
            
            website = 'http://www.maya-club.com/{}-page-{}.html'.format(urlsub[:-5],n)  #Slicing {sublist} to put "-page-{number}"
            newweb = requests.get(website)          #Getting info about the website.
            print(website)                          #Keeping track of the website url/num.
            time.sleep(2)                           #Giving a rest for a server - Good, gut bot :3
        except HTTPError:
            print("HTTPError")
        except URLError:
            print('The server could not be found!')
        
        bs = bs4.BeautifulSoup(newweb.text,"lxml")                       #Converting {newweb} into readable form of html.
        anytext = bs.find_all("div", {"class":"subtable altbg2 t_msg"})  #The imgurls are hidden within broken bbcode in <div> "subtable altbg2 t_msg" - class. 
        if len(anytext) != 0:                              # If there are msges (not empty website):
            for img in anytext:                                  # For text (img) extracted from bs.find_all
                print(img.text)                                  # For tracking print list of imgurls in each message.
                imglist += re.findall(specify, img.text)         # Add to the imglist only specified(specify-below/outside function) text based on img.text.
# Could add slicing to get urls only/without [img] tags          # In our case [img](anything)[/img].lower(). Some posts are with [IMG]/[Img] etc.
        else:                                              # If no msges (empty website)
            print("Breaking loop at{}".format(website))          # Print website/num at which we gonna break
            break
        n += 1                                          # Add +1 to thread page.

    
#//////////////////////////////////////////////////////////////////////////////////////////
#                                Some lists/arguments 
#//////////////////////////////////////////////////////////////////////////////////////////
mainlist = [] #Store subpages urls
sublist = []  #Store threads urls
imglist = []  #Store img urls
specify = re.compile("\[img\]+\S+\[/img\]", re.IGNORECASE)  #Filter to find [img](anything)[/img].lower() for storing

#//////////////////////////////////////////////////////////////////////////////////////////
#                    Main page - looking for subpages (No func)
#//////////////////////////////////////////////////////////////////////////////////////////
#Checking main webpage for subpages. And adding them into mainlist.
try:
    strona = requests.get('http://www.maya-club.com/archiver/')
except HTTPError as e:
    print(e)
except URLError as e:
    print('The server could not be found!')


soup = bs4.BeautifulSoup(strona.text,"lxml")
for url in soup.find_all("li"):
    if url.a != None:
        mainlist.append(url.a["href"])

print(mainlist)  # Keeping for futher tracking how many more left. Still possible to track.

#//////////////////////////////////////////////////////////////////////////////////////////
#                   Executing funcs and storing imgurl list in new file
#//////////////////////////////////////////////////////////////////////////////////////////
for x in mainlist:
    geturlThreadPage(x)

print(sublist)
print("Getting Thread URLs finished!")
time.sleep(10)

for x in sublist:
    geturlImgPage(x,imglist)
    
print(imglist)

with open("imglist.txt", "w") as output:
    output.write(str(imglist))
```
